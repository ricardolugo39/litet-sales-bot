```python
pd.read_sql(
    "SELECT COUNT(*) as total_rows FROM orders",
    conn,
)
```

```python
conn.execute("DROP TABLE IF EXISTS orders_clean")

conn.execute(
    """
    CREATE TABLE orders_clean AS

    SELECT *
    FROM (
        SELECT *,
               ROW_NUMBER() OVER (
                   PARTITION BY
                       [amazon-order-id],
                       asin,
                       quantity,
                       [purchase-date],
                       [product-name]
                   ORDER BY imported_at DESC
               ) as rn
        FROM orders
    )
    WHERE rn = 1
    """
)

conn.commit()
```

```python
pd.read_sql(
    "SELECT COUNT(*) as cleaned_rows FROM orders_clean",
    conn,
)
```

```python
pd.read_sql(
    """
    SELECT
        [amazon-order-id],
        asin,
        quantity,
        [purchase-date],
        [product-name],
        COUNT(*) as c
    FROM orders_clean
    GROUP BY
        [amazon-order-id],
        asin,
        quantity,
        [purchase-date],
        [product-name]
    HAVING COUNT(*) > 1
    """,
    conn,
)
```
